```python
import pandas as pd
```

```python
dev = pd.read_csv("new_vec_dev_origin.csv")[["Drug_ID", "Target_ID", "Drug", "Target", "Y"]].reset_index(drop=True)
train = pd.read_csv("new_vec_train_origin.csv")[["Drug_ID", "Target_ID", "Drug", "Target", "Y"]].reset_index(drop=True)
test = pd.read_csv("new_vec_test_origin.csv")[["Drug_ID", "Target_ID", "Drug", "Target", "Y"]].reset_index(drop=True)
```

```python
dev
```

```python
concat_df = pd.concat([dev, train, test])[["Drug_ID", "Target_ID", "Drug", "Target", "Y"]].reset_index(drop=True)
concat_df
concat_df.to_csv("vec.txt", sep= " ", index=False)
```

```python
train[["Drug_ID", "Target_ID", "Drug", "Target", "Y"]].reset_index(drop=True).to_csv("train_vec.txt", sep= " ", index=False)
dev[["Drug_ID", "Target_ID", "Drug", "Target", "Y"]].reset_index(drop=True).to_csv("dev_vec.txt", sep= " ", index=False)
test[["Drug_ID", "Target_ID", "Drug", "Target", "Y"]].reset_index(drop=True).to_csv("test_vec.txt", sep= " ", index=False)

```
